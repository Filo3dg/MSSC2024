# Spin Orbit Coupling Effect on Electronic structure

The first step as in the other tutorial is to download the CRYSTALClear package

```python
pip install crystalclear
```

We can now import the package

```python
from CRYSTALClear.crystal_io import Properties_output
import CRYSTALClear.plot as CCplt
```

Define the path of the BAND.DAT files that you are interested in plotting

```python
path = ['output/WS2_pbe_nosoc.BAND', 'output/WS2_pbe.BAND']
```

We can now read and plot the content of the files to reveil the effect that spin orbit coupling has on th electronic structure of the system

```python
bands = []

for file in path:
    bands.append(Properties_output().read_electron_band(file))

CCplt.plot_electron_band(bands, energy_range=[-5,5], linewidth=[1.2,1.2], labels=['NOSOC', 'SOC'], figsize=(10,10))
```
