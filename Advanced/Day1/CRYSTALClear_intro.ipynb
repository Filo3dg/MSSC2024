## Welcome to a Jupyter Notebook!

CRYSTALClear is an open-source python framework for the quantum chemistry CRYSTAL software.
A list of its core functionalities is reported below:
- Extraction of quantities from output files
- Band structure and DOSS
- Elastic properties (Young and shear moduli, linear compressibility, Poisson ratio)
- Harmonic and anharmonic IR and Raman spectra
- QTAIM analysis (electron density...)
- Electron transport properties

You can learn more on the [GitHub page](https://github.com/crystaldevs/CRYSTALClear). You are more than welcome to contribute!

Now, we need to import modules from CRYSTALClear:

```python
from CRYSTALClear.crystal_io import Properties_output
from CRYSTALClear import plot as CCplt
```

Specify your output file path:

```python
path = '../../Data/sio2_test_dat.BAND'
```

Let's create a CRYSTALClear object:

```python
data = Properties_output().read_electron_band(path)
CCplt.plot_electron_band(data, energy_range=[-3.5, 0.5], k_labels=['Gamma', 'M', 'K', 'Gamma'], title='hTaAs')
```

```python
CCplt.plot_electron_band(data, unit='Hartree', energy_range=[-0.3, 0.05], k_labels=['Gamma', 'M', 'K', 'Gamma'],
                         title='hTaAs', linestl='--',linewidth=2, figsize=(16,10), color='green', fermi='black', fermiwidth=5, fermialpha=0.5)
```
